```python
import pandas as pd
```

**Sales Report**

```python
df_sales = pd.read_excel('Data for the task.xlsx', sheet_name = 'Sale Report')
print("Initial Shape: ", df_sales.shape)

#drop na
df_sales.dropna(inplace = True)

print("After dropping na shape: ", df_sales.shape)

#change to correct datetime
df_sales['Date'] =  pd.to_datetime(df_sales['Date'], format='%d/%m/%Y')


#to csv
df_sales.to_csv('Sale Report.csv', index = False)

df_sales.head()
```
